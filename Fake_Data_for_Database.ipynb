```
import random as rm
rm.seed(1)

# STUDENT TABLE

# create id's using random integers
stud_id = ['{id}'.format(id = rm.randrange(10**6,10**7)) for _ in range(9)]
dept_id = ['{id}'.format(id = rm.randrange(10**4,10**5)) for _ in range(9)]

# list of first and last names
stud_fnames = ['edward','carlos','mitchell','sydney','kevin','daniel','ashley','richard']
stud_lnames = ['smith','gentile','mora','palacios','nolasco','johnson','williams','brown']

# social and dob using random integers and random samples
stud_social = ['{sn1}-{sn2}-{sn3}'.format(sn1 = rm.randrange(10**2,10**3), 
                                          sn2 = rm.randrange(10,10**2), 
                                          sn3 = rm.randrange(10**3,10**4)) for _ in range(9)]

months = ['jan','feb','mar','apr','may','june','july','aug','sep','oct','nov','dec']
stud_dob = ['{dd}-{mm}-{yy}'.format(dd = rm.randrange(1,30), 
                                    mm = rm.sample(months, 1)[0], 
                                    yy = rm.randrange(90,99)) for _ in range(9)]

# random address generator
street_names = ['baltimore','prairie','ingelwood','huntington','first','alabama','adams','beach','hawthorne']
stud_street = ['{num} {street} blvd'.format(num = rm.randrange(1,10**3), 
                                            street = rm.sample(street_names,1)[0]) for _ in range(9)]

stud_city = ['hawthorne','huntington','isla vista','goleta','manhattan beach','hermosa beach','compton','portland','santa monica']
stud_state = ['ca','co','ma','or','nv','ar','tx','sc','tn']
stud_zip = ['{num1}-{num2}'.format(num1 = rm.randrange(10**4,10**5), 
                                   num2 = rm.randrange(10**3,10**4)) for _ in range(9)]

# generating random emails
email_locations = ['cabrini.edu','google.com','yahoo.com','outlook.com']
stud_email = ['{first}{last}@{email}'.format(first = stud_fnames[ind][0], 
                                             last = stud_lnames[ind], 
                                             email = rm.sample(email_locations,1)[0]) for ind,_ in enumerate(stud_fnames)]  

# generating random phone
stud_phone  = ['{first}-{second}-{third}'.format(first = rm.randrange(10**2,10**3), 
                                                   second = rm.randrange(10**2,10**3), 
                                                   third = rm.randrange(10**3,10**4)) for _ in range(9)]
```
